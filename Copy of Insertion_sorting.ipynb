```python
#Implementing Insertion sorting

def insertion_sort(arr):
  for i in range(1,len(arr)):
    key = arr[i]
    j = i-1
    while j>=0 and arr[j]>key:
      arr[j+1] = arr[j]
      j=j-1
      arr[j+1]=key

  return arr


print(insertion_sort([21,55,1,19,2,465,72,0,88]))



#Implementing stacks

test_arr = [10]
n = len(test_arr)

def push_element(val,a,n,top):
  if top == n:
    print("No space")
  else:
    a[top]=val
    top = top+1
  return a

def pop_element(a,top):
  if top == -1:
    print("No element to remove")
  else:
    print(a[top])
    top=top-1

  return a


def display_element(a,top):
  for i in range(0,top):
    print(a[i])


```
