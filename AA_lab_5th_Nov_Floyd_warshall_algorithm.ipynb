```python
# Number of vertices
nV = 4
INF = 999

# Floyd-Warshall Algorithm
def floyd(G):
    dist = list(map(lambda p: list(map(lambda q: q, p)), G))

    # Adding vertices individually
    for r in range(nV):
        for p in range(nV):
            for q in range(nV):
                dist[p][q] = min(dist[p][q], dist[p][r] + dist[r][q])

    output_sol(dist)


# Printing the output
def output_sol(dist):
    for p in range(nV):
        for q in range(nV):
            if dist[p][q] == INF:
                print("INF", end=" ")
            else:
                print(dist[p][q], end=" ")
        print(" ")  # Move to the next line after printing each row


# Input graph matrix
G = [
    [0, 5, INF, INF],
    [50, 0, 15, 5],
    [30, INF, 0, 15],
    [15, INF, 5, 0]
]

# Run the Floyd-Warshall Algorithm
floyd(G)
```

```python
# Multistage Graph - Shortest Path using Dynamic Programming

# Total number of vertices
N = 12

# Represent infinity as a large number
INF = float('inf')

def shortest_distance(graph):
    # Initialize distance array with infinity
    distance = [INF] * N

    # Distance to the destination (last node) is 0
    distance[N - 1] = 0

    # Process nodes from N-2 down to 0
    for i in range(N - 2, -1, -1):
        # Check all possible nodes j that i can go to
        for j in range(i + 1, N):
            if graph[i][j] == INF:
                continue
            distance[i] = min(distance[i], graph[i][j] + distance[j])

    # Return the shortest distance from start (0) to end (N-1)
    return distance[0]


# Graph representation (Adjacency Matrix)
# INF means there is no direct edge between i and j
graph = [
    [INF, 9, 7, 3, 2, INF, INF, INF, INF, INF, INF, INF],  # From node 1
    [INF, INF, INF, INF, INF, 4, 2, INF, INF, INF, INF, INF],  # From node 2
    [INF, INF, INF, INF, INF, 2, 7, INF, INF, INF, INF, INF],  # From node 3
    [INF, INF, INF, INF, INF, INF, INF, 11, INF, INF, INF, INF],  # From node 4
    [INF, INF, INF, INF, INF, INF, INF, 8, INF, INF, INF, INF],   # From node 5
    [INF, INF, INF, INF, INF, INF, INF, INF, 6, 5, INF, INF],     # From node 6
    [INF, INF, INF, INF, INF, INF, INF, INF, 4, 3, INF, INF],     # From node 7
    [INF, INF, INF, INF, INF, INF, INF, INF, INF, 5, 6, INF],     # From node 8
    [INF, INF, INF, INF, INF, INF, INF, INF, INF, INF, INF, 4],   # From node 9
    [INF, INF, INF, INF, INF, INF, INF, INF, INF, INF, INF, 21],  # From node 10
    [INF, INF, INF, INF, INF, INF, INF, INF, INF, INF, INF, 5],   # From node 11
    [INF, INF, INF, INF, INF, INF, INF, INF, INF, INF, INF, INF]  # From node 12
]

# Calculate the shortest distance
print("Distance from start:", shortest_distance(graph))
```
